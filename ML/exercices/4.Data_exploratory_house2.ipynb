# **Exploratory 2**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/house_pricing_clean2.csv', sep=',')
```

```python
pd.options.display.max_columns = None   #para que muestren todas las columnas
df
```

## Bars plot

### YearRemodAdd / SalePrice

```python
df.groupby("YearRemodAdd")["SalePrice"].mean().plot(kind = "bar")
plt.xticks(rotation=45, ha='right', fontsize=6)
```

### MSZoning / SalePrice

```python
df.groupby("MSZoning")["SalePrice"].mean().plot(kind = "bar")
```

### GrLivArea / SalePrice

```python
df.groupby("GrLivArea")["SalePrice"].mean().plot(kind = "bar")
```

## Correlations

```python
df.corr()
```

### Head Map

```python
#sns.heatmap(df.corr())
fig, ax = plt.subplots(figsize=(10, 8), dpi=100)
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5, linecolor='white')

plt.show()
```

```python
## Podemos ver correlacion entre TotalBsmtSF y 1stFirSF, tambien vemos mucha correlacion entre GarageCars y GarageArea
## y ya con un poco menos pero un poco el GrLivArea y el TotRmsAbvGrd.
```

### Pairplots

```python
#sns.pairplot(df)
sns.pairplot(df, diag_kind='kde', plot_kws={'alpha': 0.6}, diag_kws={'alpha': 0.6}, markers='.')

plt.show()
```

```python
'''
En el pariplot se puede ver una grafica interesante en el YearBuilt and YearRemodAdd donde llega un momento que sube muy rapido,
tambien como hemos visto en los gráficos en el pairplot en el YearBuilt o YearRemodAdd suben con el PriceSales y tambien el 
GrLivArea con el PriceSales
'''
```
