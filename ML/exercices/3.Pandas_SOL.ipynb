## Age Bracketing with Occupations

Create a new DataFrame where you categorize users into age brackets: '18-30', '31-50', '51-70', and '70+'. For each occupation and age bracket, compute the average age, total count, and gender distribution percentage.

```python
# Categorize by age brackets
bins = [18, 30, 50, 70, 100]
labels = ['18-30', '31-50', '51-70', '70+']
df['age_bracket'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Group by occupation and age bracket
grouped = df.groupby(['occupation', 'age_bracket'])

# Compute average age, count, and gender distribution
result = grouped.agg(
    avg_age=('age', 'mean'),
    total_count=('user_id', 'size'),
    male_pct=('gender', lambda x: (x == 'M').mean().round(1) * 100),
    female_pct=('gender', lambda x: (x == 'F').mean().round(1) * 100)
).reset_index()

result
```

## Create a stacked bar chart to visualize the distribution of males and females in each occupation.

```python
gender_occupation = df.groupby(['occupation', 'gender']).size().unstack().fillna(0)
gender_occupation.plot(kind='bar', stacked=True, color=['lightseagreen', 'lightgreen'])
plt.title('Gender Distribution by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Users')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Gender Dominance by Zip Code

Identify zip codes where a single gender represents more than 90% of individuals.

```python
grouped_zip = df.groupby(['zip_code', 'gender']).size().unstack(fill_value=0)
grouped_zip['total'] = grouped_zip['M'] + grouped_zip['F']
grouped_zip['M_pct'] = grouped_zip['M'] / grouped_zip['total'] * 100
grouped_zip['F_pct'] = grouped_zip['F'] / grouped_zip['total'] * 100

grouped_zip[(grouped_zip['M_pct'] > 90) | (grouped_zip['F_pct'] > 90)]
```

## Most Diverse Occupations

Identify the top 5 occupations that have the closest male-to-female ratio.

```python
grouped_occ = df.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
grouped_occ['total'] = grouped_occ['M'] + grouped_occ['F']
grouped_occ['M_ratio'] = grouped_occ['M'] / grouped_occ['total']
grouped_occ['F_ratio'] = grouped_occ['F'] / grouped_occ['total']
grouped_occ['diff'] = abs(grouped_occ['M_ratio'] - 0.5)

grouped_occ.nsmallest(5, 'diff')
```

## User Density by State

Assuming the first two characters in the `zip_code` represent a USA state, find out which state has the highest user density (number of users per unique occupation).

```python
df['state'] = df['zip_code'].str[:2]

state_counts = df.groupby('state')['user_id'].nunique()
occupation_counts = df.groupby('state')['occupation'].nunique()

density = state_counts / occupation_counts
density.idxmax()
```

## Pie Gender Ratio

Based on the age bracketing in exercise 2.1., for each age bracket, plot a pie chart of the gender ratio side-by-side in a 2x2 grid.

```python
bins = [18, 30, 50, 70, 100]
labels = ['18-30', '31-50', '51-70', '70+']
df['age_bracket'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

fig, axs = plt.subplots(2, 2, figsize=(10,10))

for ax, label in zip(axs.ravel(), labels):
    data = df[df['age_bracket'] == label]['gender'].value_counts()
    ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=['lightseagreen', 'lightgreen'])
    ax.set_title(f'Gender Ratio for {label}')

plt.tight_layout()
plt.show()
```

## Occupations Heatmap

Create a heatmap where the x-axis represents the top 10 occupations, the y-axis represents the top 10 zip codes (by user count), and the cell color indicates the number of users.

```python
import seaborn as sns

top_10_occupations = df['occupation'].value_counts().head(10).index
top_10_zipcodes = df['zip_code'].value_counts().head(10).index

filtered_df = df[df['occupation'].isin(top_10_occupations) & df['zip_code'].isin(top_10_zipcodes)]
pivot_table = pd.pivot_table(filtered_df, index='zip_code', columns='occupation', values='user_id', aggfunc='count', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, linewidths=.5)
plt.title('User Density by Occupation and Zip Code')
plt.show()
```

## Occupation Age Range

For each occupation, compute the age range (max - min). Display the top 10 occupations with the most diverse age ranges using a horizontal bar chart.

```python
df['age_range'] = df.groupby('occupation')['age'].transform(lambda x: x.max() - x.min())
top_diverse_occupations = df.drop_duplicates('occupation').nlargest(10, 'age_range')

top_diverse_occupations.set_index('occupation')['age_range'].sort_values().plot(kind='barh', color='mediumpurple')
plt.title('Occupations with Most Diverse Age Ranges')
plt.xlabel('Age Range')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```
