## Gender Dominance by Zip Code

Identify zip codes where a single gender represents more than 90% of individuals.

```python
df4 = pd.read_csv('data/occupation.csv', sep='|')

gen_count = df4.groupby("zip_code")["gender"].value_counts()
total = df4.groupby("zip_code")["gender"].count()
percen = gen_count / total * 100
result = percen[percen > 90].index.get_level_values("zip_code").unique()
result
```

## Most Diverse Occupations

Identify the top 5 occupations that have the closest male-to-female ratio.

```python
gen_count = df4.groupby("occupation")["gender"].value_counts()
total = df4.groupby("occupation")["gender"].count()
percen = gen_count / total * 100
close_ratio = abs(percen - 50).sort_values().index.get_level_values("occupation").unique()
close_ratio[:5]
#close_ratio.head(10)
```
