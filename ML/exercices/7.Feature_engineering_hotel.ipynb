### One-hot encoding

```python
fts_to_onehot = [
    'hotel',
    'arrival_date_month',
    'meal',
    'country',
    'market_segment',
    'distribution_channel',
    'reserved_room_type',
    'assigned_room_type',
    'deposit_type',
    'customer_type',    
]
```

```python
# Number of unique elements, to make sure we don't create too many columns
for ft in fts_to_onehot:
    print(ft, df[ft].nunique())
```

```python
df = pd.get_dummies(df, columns=fts_to_onehot)
df.head()
```
