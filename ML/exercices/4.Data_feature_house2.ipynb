# **Featuring**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('data/house_pricing_clean2.csv', sep=',')
```

```python
pd.options.display.max_columns = None   #para que muestren todas las columnas
df
```

## Deleting Id Column

```python
df.drop("Id", axis=1, inplace=True)
df
```

## One Hot Encoded

```python
one_hot_encoded = pd.get_dummies(df["MSZoning"]) 
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("MSZoning", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["HouseStyle"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("HouseStyle", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["RoofStyle"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("RoofStyle", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Foundation"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Foundation", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Heating"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Heating", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["CentralAir"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("CentralAir", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Electrical"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Electrical", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["GarageType"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("GarageType", axis=1, inplace=True)

df
```

 ## Ordinal Encoded

```python
encoder = OrdinalEncoder(categories=[["Gtl", "Mod", "Sev"]])
df['LandSlope'] = encoder.fit_transform(df['LandSlope'].values.reshape(-1, 1))    # el reshape es para que tenga 2 dimensiones


encoder2 = OrdinalEncoder(categories=[["Y", "P", "N"]])
df['PavedDrive'] = encoder2.fit_transform(df['PavedDrive'].values.reshape(-1, 1)) 

encoder3 = OrdinalEncoder(categories=[["0", "Unf", "RFn", "Fin"]])
df['GarageFinish'] = encoder3.fit_transform(df['GarageFinish'].values.reshape(-1, 1))

pd.set_option('display.max_rows', 10)   # si pones donde esta el 10 se puede ver todas las filas, y despues las puedes cambiar a menos con un numero
df
```
