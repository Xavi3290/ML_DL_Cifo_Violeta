# **Featuring**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/house_pricing_clean.csv', sep=',')
```

```python
pd.options.display.max_columns = None   #para que muestren todas las columnas
df
```

##  Deleting Id Column

```python
df.drop("Id", axis=1, inplace=True)
df
```

## One Hot Encoded

```python
one_hot_encoded = pd.get_dummies(df["MSZoning"]) 
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("MSZoning", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Street"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Street", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Utilities"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Utilities", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["LandSlope"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("LandSlope", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["HouseStyle"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("HouseStyle", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["RoofStyle"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("RoofStyle", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Foundation"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Foundation", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Heating"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Heating", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["CentralAir"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("CentralAir", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["Electrical"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("Electrical", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["GarageType"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("GarageType", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["GarageFinish"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("GarageFinish", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["PavedDrive"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("PavedDrive", axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(df["MiscFeature"])
df = pd.concat([df, one_hot_encoded], axis=1)
df.drop("MiscFeature", axis=1, inplace=True)


df
```
