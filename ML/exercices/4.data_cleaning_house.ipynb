## Change nulls

### Electical

```python
df[df["Electrical"].isna()].T
```

```python
df["Electrical"].value_counts()
```

```python
df.loc[1379, "Electrical"] = "Sbrkr"   #The only nun change for the most value
df.loc[1379]
```

### GarageType

```python
df[df["GarageType"].isna()].T
```

```python
df["GarageType"].value_counts()
```

```python
df.loc[df["GarageType"].isna(), "GarageType"] = 0       #0 ya que no hay 
df["GarageType"].value_counts()
```

### GarageYrBlt and GarageFinish

```python
df.loc[df["GarageYrBlt"].isna(), "GarageYrBlt"] = -1    #Le pongo -1 para representar que no hay nada
df.loc[df["GarageFinish"].isna(), "GarageFinish"] = 0   #0 ya que no se a hecho
```

```python
df["GarageYrBlt"].value_counts()
df["GarageFinish"].value_counts()
```

### MiscFeature

```python
df["MiscFeature"].isna().value_counts()
```

```python
df[df["MiscFeature"].notna()]
df["MiscFeature"].value_counts()
```

```python
df.loc[df["MiscFeature"].isna(), "MiscFeature"] = 0  #0 porque no tienen 
pd.options.display.max_columns = None   #para que muestren todas las columnas
df
```
